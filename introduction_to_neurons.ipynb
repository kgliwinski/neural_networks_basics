## Inspiration. Human neurons



Neurons gather signals from predecessors (other neurons) - they can travel with different speeds, so we will call them **weights**.

When signals get into the neuron (sum) and multiplied by weights they will reach some treshold, the neuron will send signal further.

## Competition. Traditional computer vs Neural net.

Traditional computer:
- one very sophisticated computer     
- warking at high speed/frequency (GHz)
- Quite powerful overall $ 10^{11} op./sec.$ - Intel I7
- Acquires knowledge by:
    - being programmed
    - modeling a known algorithm: clear, exact, completely specified
- best for:
    - clear and exactly defined tasks
- very high accuracy of calculations
- low fault tolerance
- knowledge stored in program, code and algorithms


Neural net:
- many very simple processors

    - things we do unconciously
- rather low accuracy of calculations
- high fault tolerance
- knowledge stored in weights

## Neuron. Structure, operation, properties



- variable $u = w_1 x_1 + w_2 x_2$ - weighted sum
- activation function $y = f(u)$
- treshold is also a parameter of neuron, can be treated as one of the weights $w$



### Sigmoid function

- Activation function: $y = f(u) = \frac{1}{1 + \exp^{\Beta u}}$
- derivative: $f'(u) = \Beta \cdot y(1-y)$
- Unipolar sigmoid: $y = \frac{1}{1 + \exp^{\Beta u}} = f_u(u)$
- Bipolar sigmoid: $y = 2\cdot y - 1 = \tanh (\frac{\Beta}{2}) = f_b(u)$

### Sigmoid neuron: vector notation

- $\bar{x} = [x_1, x_2, ... , x_S]_{S \times 1}$
- $\bar{x}^{1} = [-1 = x_0, x_1, x_2, ... , x_S]_{S \times 1}$
- $\bar{w}^{1} = [w_0, w_1, w_2, ... , w_S]_{S \times 1}$

## Vapnik-Chervonenkis dimension. VC-dim

Measurement of the learning capacities of a learning machine. It coincides with the complexity of a model, the number of weights.

- dychotomy: distinction/separation into 2 classes - particular case of classification
- VCdim - cardinality of the biggest set of points in input space, for which the learning machine (neuron)
    - is able to provide all possible dichotomies
    - is able to classify them in all possible ways
